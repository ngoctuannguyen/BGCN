```python
import pandas as pd
import os
```

```python
# Bundle Item
bi_DF = pd.read_csv('./bundle_item.csv')
bi_DF
```

```python
def mapping_item_id(item_id):
    item_id = sorted(item_id)
    item_id_map = dict()
    for i in range(len(item_id)):
        item_id_map[item_id[i]] = i
    return item_id_map
```

```python
item_id = []
for i in range(bi_DF.shape[0]):
    item_id.append(bi_DF.iloc[i]['item ID'])
item_id_mapping = mapping_item_id(item_id)
print(item_id_mapping)
```

```python
new_path = './bundle_item.txt'
if os.path.exists(new_path):
    os.remove(new_path)

bundle = []
item = []

with open(new_path, 'a') as f:
    for i in range(bi_DF.shape[0]):
        first_col = bi_DF.iloc[i]['bundle ID']
        sec_col = bi_DF.iloc[i]['item ID']
        bundle.append(bi_DF.iloc[i]['bundle ID'])
        item.append(bi_DF.iloc[i]['item ID'])
        #print(bi_DF.iloc[i]['bundle ID'])
        f.write(f'{first_col}\t{sec_col}\n')
```

```python
# User Item
ui_DF = pd.read_csv('./user_item.csv')
ui_DF = ui_DF.drop(columns=['timestamp'])
ui_DF
```

```python
new_path = './user_item.txt'
if os.path.exists(new_path):
    os.remove(new_path)

with open(new_path, 'a') as f:
    for i in range(ui_DF.shape[0]):
        first_col, sec_col = ui_DF.iloc[i]['user ID'], ui_DF.iloc[i]['item ID']
        f.write(f'{first_col}\t{sec_col}\n')
```

```python
# U-B
ub_DF = pd.read_csv('./user_bundle.csv')
ub_DF = ub_DF.drop(columns=['timestamp'])
ub_DF
```

```python
user_id = []
bundle_id = []

for i in range(ub_DF.shape[0]):
    first_col = ub_DF.iloc[i]['user ID']
    sec_col = ub_DF.iloc[i]['bundle ID']
    user_id.append(first_col)
    bundle_id.append(sec_col)
print(user_id)
print(bundle_id)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test_val, Y_train, Y_test_val = train_test_split(user_id, bundle_id, test_size=0.2)
X_test, X_val, Y_test, Y_val = train_test_split(X_test_val, Y_test_val, test_size=0.5)

type_of_learning = ['train', 'test', 'tune']
for type in type_of_learning:
    new_path = f'./user_bundle_{type}.txt'
    if os.path.exists(new_path):
        os.remove(new_path)
    with open(new_path, 'a') as f:
        if type == 'train':
            for i in range(len(X_train)):
                f.write(f'{X_train[i]}\t{Y_train[i]}\n')
        elif type == 'test':
            for i in range(len(X_test)):
                f.write(f'{X_test[i]}\t{Y_test[i]}\n')
        else:
            for i in range(len(X_val)): 
                f.write(f'{X_val[i]}\t{Y_val[i]}\n')
```

```python
number_of_users = len(set(user_id))
number_of_bundle = len(set(ub_DF['bundle ID']))
number_of_item = len(set(ui_DF['item ID']))
print(number_of_users, number_of_bundle, number_of_item)
new_path = './electronic_data_size.txt'
if os.path.exists(new_path):
    os.remove(new_path)
with open(new_path, 'a') as f:
    f.write(f'{number_of_users}\t{number_of_bundle}\t{number_of_item}')
```
